# Python Data Types for Beginners

In Python, every piece of data is an object. Understanding Python’s built-in data types is fundamental for writing efficient, readable, and bug-free programs. This notebook will introduce you to Python’s core data types and show you best practices on how to work with them.

## 1. Numeric Types

Python has three main numeric data types:

### 1. Integer (`int`)  
Whole numbers without a fractional component.  
**Examples:** `0`, `1`, `-5`, `999`, `42`, `-10`

### 2. Float (`float`)  
Numbers with decimal points or in exponential form.  
**Examples:** `3.14`, `2.7e3`, `-0.001`

### 3. Complex (`complex`)  
Numbers with both real and imaginary parts.  
**Examples:** `2+3j`, `-1.5+2.7j`


## 1.1 Integers
Definition: Integers are whole numbers without a fractional component.
Examples: `-10, 0, 42`

````python
# Example of integers
num1 = 50
num2 = -10
num3 = 0

print('num1 ->', num1, type(num1))
print('num2 ->', num2, type(num2))
print('num3 ->', num3, type(num3))
````

### Best Practices:
1. Use descriptive variable names: Instead of x or y, use user_age or items_count.

2. Handle potential overflow: In Python 3, integers are of unlimited size (limited by available memory), so you rarely need to worry about overflow.

3. Avoid ambiguous naming: For example, `O` (capital letter O) or `l` (lowercase L) can be confused with numbers.

## 1.2 Floats

Definition: Floats represent real numbers that include a fractional component.
Examples: 3.14159, -0.001, 1.5e2 (which equals 150.0).

````python
# Example of floats
pi = 3.14159
small_float = -0.001
expo_float = 1.5e2 # 150.0 in decimal

print('pi is ->', pi, type(pi))
print('small_float is ->', small_float, type(small_float))
print('expo_float is ->', expo_float, type(expo_float))
````

### Best Practices:
1. Watch out for floating-point precision issues. Use decimal or fractions modules if you need precise decimal arithmetic (e.g., currency calculations).
2. Use underscores for large numbers: Python 3.6+ allows underscores in numeric literals for readability, e.g., 1_000_000.0 instead of 1000000.0.
3. Check your computations: Floating-point arithmetic can lead to subtle bugs in equality checks. Prefer a tolerance-based approach if comparing floats, for example:

### Example of the Problem

```python
a = 0.1 + 0.2
print(a)
```

You expect:

```
0.3
```

But Python actually stores:

```
0.30000000000000004
```

Now if you check:

```python
print(a == 0.3)
```

Output:

```
False
```

😲 Why False?

Because:

```
0.30000000000000004  ≠  0.3
```

Even though they look the same to us!

This is called a **floating-point precision issue**.

---

#### What is a Tolerance-Based Approach?

Instead of checking:

```python
a == 0.3
```

We check if the difference is **very small**.

Example:

```python
a = 0.1 + 0.2

print(abs(a - 0.3) < 0.00001)
```

Output:

```
True
```

#### What Is Happening Here?

`abs(a - 0.3)`
→ Finds the difference between numbers.

If the difference is **very small (close to zero)**,
we treat them as equal.

This small allowed difference is called **tolerance**.

#### Even Better Way (Recommended)

Python has a built-in function:

```python
import math

a = 0.1 + 0.2

print(math.isclose(a, 0.3))
```

Output:

```
True
```

This is the safest and cleanest way.

#### Simple Rule to Remember

❌ Don’t compare floats using `==`
✅ Compare floats using a small tolerance
✅ Or use `math.isclose()`


### Final Summary

* Decimal numbers are not stored perfectly in computers.
* Small hidden differences can appear.
* `==` may fail even when numbers look equal.
* Always use tolerance when comparing floats.


````python
b = 0.1 + 0.2
print(b)
````

````python
print(b == 0.3)
````

````python
c = 0.1 + 0.2

print(abs(c - 0.3) < 0.00001)
````

````python
import math

d = 0.1 + 0.2

print(math.isclose(d, 0.3))
````

## 2. Strings

Definition: A string is a sequence of characters. In Python, strings can be enclosed in single quotes ('...'), double quotes ("..."), or triple quotes ('''...''' or """...""") for multi-line strings.

Examples: "Hello World", 'Python is great!', """Triple quoted strings can span multiple lines""".

````python
# Example of Strings
greeting = "Hello, World"
quote = 'I can use single quotes too'
multiline_text = '''This is a 
Multi line string.'''

print(greeting, type(greeting))
print(quote, type(quote))
print(multiline_text, type(multiline_text))
````

### String Operations

````python
# Concatination
first_name = "Muhammad"
last_name = "Abid"

full_name = first_name +" "+ last_name # Muhammad Abid

# Repetition
star_ten_time = "*" * 10

# Indexing and Slicing
first_chr = first_name[0] # M
substring = greeting[0:5] # hello


print('full_name:', full_name)
print('Repetition:', star_ten_time)
print('Firts Char:', first_chr)
print('Substring:', substring)
````

### Best Practices:

1. Use f-strings (formatted strings) for readability (Python 3.6+)

````python
Age = 25
print(f'My age is -> ', Age)
print(f"My age is: {Age}")
````

## 3. Boolean Type

Definition: Booleans are True or False values.
Use: Typically used in condition checks, loops, or logical evaluations.

````python
is_python_easy = True
like_office = False

print("Is python easy?:", is_python_easy)
print("Like an Office?:", like_office)
print('\n',type(like_office))
````

### Best Practices:

1. Use descriptive names: Instead of flag or status, use is_active or has_errors.
2. Avoid comparing booleans to True or False directly
3. Keep booleans purely boolean: Resist the temptation to mix them with numbers. Although True behaves like 1 and False like 0, it makes the code less clear.

````python
# Instead of: is_python_easy? == True
# Do this

if is_python_easy:
    print('yes python is easy to learn.')
````

### Instead of writing this:

```python
if is_python_easy == True:
    print("yes python is easy to learn.")
```

Here you are explicitly checking whether `is_python_easy` equals `True`.

#### Pythonic way (recommended):

```python
if is_python_easy:
    print("yes python is easy to learn.")
```

#### Why?

In Python, if `is_python_easy` is already a **Boolean value** (`True` or `False`), you don’t need to compare it to `True`.

Python automatically checks:

* If it’s `True` → run the code
* If it’s `False` → skip it

So:

```python
if is_python_easy:
```

already means:

```python
if is_python_easy == True:
```

#### Example

```python
is_python_easy = True

if is_python_easy:
    print("yes python is easy to learn.")
```

Output:

```
yes python is easy to learn.
```

If:

```python
is_python_easy = False
```

Nothing prints.

#### Why this is better

* Cleaner
* Shorter
* More readable
* Standard Python style

#### Bonus Tip

Similarly, instead of:

```python
if is_python_easy == False:
```

Write:

```python
if not is_python_easy:
```

#### Rule of Thumb

If a variable is already a Boolean, **don’t compare it to `True` or `False`** — just use it directly.

## 5. Collections Overview

While primitive data types (like int, float, bool, and str) hold single values, Python’s built-in collections let you group multiple items in flexible ways. The four primary built-in collections are:

1. List – An ordered, mutable sequence.
2. Tuple – An ordered, immutable sequence.
3. Dictionary – An unordered set of key-value pairs.
4. Set – An unordered collection of unique elements.

Below, we’ll explore each in depth with code examples and best practices.

### 5.1 List

Definition & Characteristics

1. A list is created using square brackets `[]` or the `list()` constructor.
2. It is ordered, meaning elements retain the order in which they’re inserted.
3. It is mutable: You can add, remove, or modify elements after creation.

````python
# Creating list in different way
empty_list = []
numbers = [1,2,34,56,7,89,0]
mixed_list = ["Hello", 123, 5.67, True]

print('This is an empty list -> ', empty_list)
print('List of numbers -> ', numbers)
print('Different data types in Mixed list -> ', mixed_list)

# Accessing elements by index
first_number = numbers[0] # 1
last_item = mixed_list[-1]
print('\nFirst number is -> ', first_number)
print('Last item is -> ', last_item)
````

**Adding, Removing and Modifying Elements**

````python
# Adding elements
numbers.append(70)       # Adds 70 at the end
numbers.insert(2, 10)   # Inserts 10 at index 2

# Removing elements
removed_item = numbers.pop()      # by default delete last item
numbers.remove(10)                # delete specified item

# Modifying an element
numbers[1] = 72                # Change the element at index 1 to 72

print("Modified numbers list ", numbers)
print("Removed item:", removed_item)
````

### 5.2 Tuple

Definition & Characteristics

1. A tuple is created using parentheses () or the tuple() constructor.
2. It is ordered, similar to a list.
3. It is immutable: You cannot add, remove, or change elements once defined.
Creation & Basic Operations

````python
# Creating tuples
empty_tuple = ()
single_element_tuple = (11,)
numbers_tuple = (10, 30, 50, 70, 90)
mixed_tuple = ('Hello', 25, True)

print('This is an Empty tuple -> ', empty_tuple)
print('This is single element tuple ->', single_element_tuple)
print('Numbers tuple:', numbers_tuple)
print('Mixed tuple:', mixed_tuple)

# Accessing elements by index
first_item = numbers_tuple[0]    # 10
print('\nFirst item is ->', first_item)

# Slicing works similarly to lists
sub_tuple = numbers_tuple[2:]      # 50, 70, 90
print('Slicied sub-tuple is this ->', sub_tuple)
````

**Immutability**

````python
# Trying to modify a tuple will raise an error
try:
    numbers_tuple[1] = 100   # This is not allowed
except TypeError as e:
    print('Error:', e)
````

````python
# Unpacking a tuple in a seprate veriable
person_tuple = ("Abid", 30, "Engineer")
name, age, job = person_tuple  # Unpack in order

print('Name:', name)
print('Age:', age)
print('Job:', job)
````

Best Practices for Tuples

1. Use tuples for fixed collections of items that don’t change, like coordinates or constant data.
2. Leverage tuple unpacking to write cleaner code when returning multiple values from a function.
3. Don’t modify: If you need to frequently alter the collection, consider a list.

### 5.3 Dictionary

Definition & Characteristics

1. A dictionary is created using curly braces {} with key-value pairs, or using dict().
2. It is unordered (though Python 3.7+ maintains insertion order as an implementation detail).
3. Keys must be unique and immutable (e.g., strings, numbers, tuples). Values can be anything.
I4. t is mutable: You can add, remove, or change key-value pairs.

**Creation & Basic Operations**

````python
# Creating dictioneries
empty_dict = {}
person_dict = {'name': 'Abid', 'age': 30}
another_dict = {'city': 'Nawabshah', 'country': 'Pakistan'}

print('Empty dictionary is ->', empty_dict)
print('Person dictionary is ->', person_dict)
print('Another dictionary is ->', another_dict)

# Accessing value by key
print('Name:', person_dict['name'])
print('City:', another_dict['city'])
print('Pincode:', another_dict.get('pin'))  # .get() is safer, returns None if key is't found.
````

**Adding, Updating, and Removing Keys**

````python
# Adding or updating a key-value paires
person_dict['City'] = 'Lahore' # Adds city if it dosent exist, else updates it.
print('Updated Person dictionary is this ->', person_dict)

# Removing a key-value pair
removed_value = person_dict.pop('age')  # Removed age and returns it's value.
print('Removed Age is ->', removed_value)

# Chacking if a key exists
if 'country' in another_dict:
    print('Country key is present')
````

````python
print('Keys:', another_dict.keys())
print('Vlues:', another_dict.values())
print('Items:', another_dict.items())
````

Best Practices for Dictionaries

1. Use dictionaries when you need fast lookups by key.
2. Keys should be immutable and descriptive (e.g., strings that convey meaning).
3. Use .get(key, default) to avoid KeyError when a key might not exist.
4. Dictionaries are great for configuration data, or bundling related data attributes without creating a custom class.

### 5.4 Set

Definition & Characteristics

1. A set is created using curly braces {} or the set() constructor.
2. It is unordered and does not allow duplicate elements.
3. It is mutable: You can add or remove elements.


Creation & Basic Operations

````python
# Creating sets
empty_set = set()
numbers_set = {2,1,4,1,2} # Duplicates are removed automatically
mixed_set = {'Hello', 3, 2, 1, 5.3, 2.1, (33,22,11,00)}

print('Empty set is:', empty_set)
print('Set of numbers is:', numbers_set)
print('Different data types in set:', mixed_set)
````

**Adding and Removing Elements**

````python
print('Before updating set:', numbers_set)
numbers_set.add(3) # Adds 3 to the set
numbers_set.discard(5)  # Remove 5 if it exists, no error if it dosen't
#numbers_set.remove(5) # Removes 5 but raise a KeyError if it dosen't exist

print('After updating set:', numbers_set)
````

### Set Operations

Sets support mathematical set operations like union, intersection, difference, etc.

````python
set_a = {2,1,4,3}
set_b = {4,3,6,5}

# Union: all elements from both sets without duplicate values
print('Union:', set_a.union(set_b)) # or set_a | set_b

# Intersection: elements that common in both sets
print('Intersection:', set_a.intersection(set_b))  # or set_a & set_b

# Difference: elements in set_a but not in set_b
print('Difference:', set_a.difference(set_b)) # or set_a - set_b

# Symmetric Difference: elements in either set_a or set_b not both meant (unique values from both sets)
print('Symmetric Difference:', set_a.symmetric_difference(set_b)) # or set_a ^ set_b
````

**Best Practices for Sets**

1. Use sets when you need to efficiently test for membership (in check) and remove duplicates.
2. Avoid storing mutable items in a set—items must be hashable (immutable).
3. Use set operations for cleaner and more efficient code when dealing with membership and overlaps.
