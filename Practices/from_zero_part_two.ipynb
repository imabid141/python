### 🎯 Your Next Challenge

Upgrade your game so it:

1. Generates a random number **1–10**
2. Lets the user **keep guessing**
3. Tells them:

```
Too high
Too low
Correct
```

Example:

```
Guess number: 4
Too low

Guess number: 9
Too high

Guess number: 7

```

````python
# Small Challenge

import random
number = random.randint(1,10)
guess = int(input("Guess number: "))

while guess != number:
    if guess > number:
        print("Too high")
    else:
        print("Too low")
    guess = int(input("Guess again: "))
    
print("Correct!")
print("The number was:", number)
````

#### 🎯 Your Next Mini Mission

Modify the game so it also prints:

```
Attempt 1 of 3
Attempt 2 of 3
Attempt 3 of 3
```

Example:

```
Attempt 1 of 3
Guess: 5
Too low
```

When you finish, send me your new code.

````python
# Next Mini Mission

import random
jackpot = random.randint(1,10)

for attempts in range(3):
    guess = int(input("Guess the jackpot: "))
    if guess == jackpot:
        print("Correct!")
        break
    elif guess > jackpot:
        print("Guess:", guess, "\nToo high")
    else:
        print("Guess:", guess, "\nToo low")

    print(f"Attempts {attempts+1} of: 3")

else:
    print("Game Over")

print("The jackpot was:", jackpot)
````

Now your **exam question**:

Modify your program so it also prints:

```
Player choice: rock
Computer choice: scissors
Result: You win
```

Example output:

````python
# Next Exam (Teacher Mode)
import random

choices = ["rock", "scissors", "paper"] # list
computer = random.choice(choices)

player = input("Pick one: rock, scissors or paper: ")

# Now start logic
if player == computer:
    print("Draw")
elif (
    (player == "rock" and computer == "scissors") or
    (player == "scissors" and computer == "paper") or
    (player == "paper" and computer == "rock")
):
    print("Result: You win")
else:
    print("Result: Computer win")

# final
print("Computer chose:", computer)
print("Player chose:", player)
````
