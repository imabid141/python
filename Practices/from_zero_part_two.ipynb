### 🎯 Your Next Challenge

Upgrade your game so it:

1. Generates a random number **1–10**
2. Lets the user **keep guessing**
3. Tells them:

```
Too high
Too low
Correct
```

Example:

```
Guess number: 4
Too low

Guess number: 9
Too high

Guess number: 7

```

````python
# Small Challenge

import random
number = random.randint(1,10)
guess = int(input("Guess number: "))

while guess != number:
    if guess > number:
        print("Too high")
    else:
        print("Too low")
    guess = int(input("Guess again: "))
    
print("Correct!")
print("The number was:", number)
````
