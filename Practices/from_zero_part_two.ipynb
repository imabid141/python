#### Number Guessing Game

Program logic:

```
Computer picks a random number
↓
User guesses
↓
Program tells:
too high
too low
correct
```

Example:

```
Guess number: 5
Too low

Guess number: 8
Too high

Guess number: 7
Correct!
```

This teaches:

* loops
* conditions
* random numbers
* game logic
