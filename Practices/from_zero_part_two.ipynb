````python
# Next Mini Mission

import random
jackpot = random.randint(1,10)

for attempts in range(3):
    guess = int(input("Guess the jackpot: "))
    if guess == jackpot:
        print("Correct!")
        break
    elif guess > jackpot:
        print("Guess:", guess, "\nToo high")
    else:
        print("Guess:", guess, "\nToo low")

    print(f"Attempts {attempts+1} of: 3")

else:
    print("Game Over")

print("The jackpot was:", jackpot)
````

````python
# Simple To-Do list (Version 2)

tasks = []
while True:
    print("\n1. Add task")
    print("2. View task")
    print("3. Delete task")
    print("4. Exit")

    choice = input("Enter your choice")

    if choice == "1":
        task = input("Add your task: ")
        tasks.append(task)

        # Task sving
        with open("tasks.txt", "a") as file:
            file.write(task + "\n")

    elif choice == "2":
        if not tasks:
            print("No task available! ")
        else:
            for i, task in enumerate(tasks):
                print(i + 1, "-", task)

    elif choice == "3":
        if task in tasks:
            tasks.remove(task)
        else:
            print("Task not found")

    elif choice == "4":
        print("Goodby!")
        break

    else:
        print("Invalid operation")
````
