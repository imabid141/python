#### 🔎 Bonus Challenge Review (Even / Odd)

Your code:

```python

if user_num %2 == 0:
    print("Odd number")
else:
    print("Even number")
```

You made a **small logical mistake**:

`user_num % 2 == 0` means the number is EVEN.

But you printed `"Odd number"`.

#### Correct Version:

```python
user_num = int(input("Enter your number: "))
if user_num % 2 == 0:
    print("Even number")
else:
    print("Odd number")
```

#### Why?

* `%` gives remainder
* Even numbers divided by 2 have remainder 0
* Odd numbers have remainder 1

Example:

* `8 % 2 = 0` → Even
* `7 % 2 = 1` → Odd

### 🚀 Step 10: Lists (Super Important)

Lists store **multiple values in one variable**.

Example:

```python
numbers = [1, 2, 3, 4, 5]
```

Another example:

```python
fruits = ["apple", "banana", "orange"]
```

You can access items by **index**.

```python
fruits = ["apple", "banana", "orange"]

print(fruits[0])
print(fruits[1])
```

Output:

```
apple
banana
```

Important rule:

```
Index starts at 0
```

| Index | Value  |
| ----- | ------ |
| 0     | apple  |
| 1     | banana |
| 2     | orange |
