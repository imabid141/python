### 1️⃣ What Is a Function?

A **function** is a block of code that performs a specific task and can be reused.

Instead of writing the same code many times, we **put it inside a function**.

#### Example (without function)

````python
# Repatitive code

print(5 * 5)
print(6 * 6)
print(7 * 7)
````

````python
# Example (with function)

def square(x):
    return x * x

print(square(5))
print(square(6))
print(square(7))
````

### 3️⃣ Key Parts of a Function

#### 1️⃣ `def`

Tells Python **we are creating a function**.

```
def greet():
```

#### 2️⃣ Function Name

```
greet
add
square
clean_data
```

Choose meaningful names.

#### 3️⃣ Parameters

These are **inputs** to the function.

```python
def add(a, b):
```

Here:

```
a
b
```

are parameters.

#### 4️⃣ `return`

`return` sends the result back.

```python
return a + b
```

Without `return`, the function gives **None**.


#### 5️⃣ Very Important Difference

````python
# Function with return

def multiply(a, b):
    return a * b

result = multiply(3, 4)
print(result)
````

````python
# Function without return

def multiply(a, b):
    print(a * b)

result = multiply(3, 4)
print(result)
````

Output:

```
12
None
```

Because the function **did not return anything**.
