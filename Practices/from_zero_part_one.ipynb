You made a **small logical mistake**:

```python

```

#### Why?

* `%` gives remainder
* Even numbers divided by 2 have remainder 0
* Odd numbers have remainder 1

Example:

* `8 % 2 = 0` → Even
* `7 % 2 = 1` → Odd

#### Small Challenge Review (Append Mode)

Your code:

```python

task = input("Enter your task")
file.write(task)
file.close()
```

Works correctly.

But better formatting:

```python
file.write(task + "\n")
```

Otherwise tasks may appear in one line.

````python
# Challenge (Your Firs Real Mini Project)
my_book = {}

for i in range(5):
    name = input("Enter name: ")
    phone = input("Enter phone: ")
    my_book[name] = phone

    file = open("contacts.txt", "a")
    file.write(name + " : " + phone + "\n")
    file.close()

search = input("Enter name to find phone number: ")

if search in my_book:
    print("Phone number:", my_book[search])
else:
    print("Invalid user name: ")

print()

for name, phone in my_book.items():
    print(name, ":", phone)
````
